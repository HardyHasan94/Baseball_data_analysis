```python
# Q1. Which team has won most awards
# Q2. Players with high salary same as players with many awards

# feature engineering:
# fill missing values or drop columns
# categorical or numerical values, handle them
```

```python
# Q1, finding teams and players with most number of awards

# first grouping by players
players_grouped = award.groupby('playerID').groups

# sorting players by the number of times they appear in the list of awards, kv[1] is that list 
sorted_players_grouped = sorted(players_grouped.items(), key = lambda kv:len(kv[1]), reverse = True)

# defining functions to return the rows in salary df of the top 3 players
def top3_data(player):
    print('This a small award table consisting of rows for player:', player, '\n','\n', award[award['playerID']==player])
    print('\n')
    print('This a small salary table consisting of rows for player:', player, '\n','\n', salary[salary['playerID']==player])
```

```python
# 1st player
player = sorted_players_grouped[0][0]
top3_data(player)
```

```python
# 2nd player, although it has no data for salary table, most probably because this player is old

player = sorted_players_grouped[1][0]
top3_data(player)
```

```python
# 3rd player, although it has no data for salary table, most probably because this player is old

player = sorted_players_grouped[2][0]
top3_data(player)
```

# 4 Conclusion & Communication

```python
# function for plotting

def plotter(year,salary, playerID):
    fig,ax = plt.subplots(figsize = (10,5))
    plt.plot(year, salary, 'bo')
    ax.set(xlabel = 'Years', ylabel = 'Salary', title = playerID)
    plt.grid(ax)
    plt.show()
```

```python
# From the exploration above we can see that the 1st player's salary almost doubles for each year, 
# sometimes even higher due to the number of awards he receives that year.
# Here is a plot of his salary over the years active as player

years = salary[salary['playerID']=='bondsba01']['yearID']
salaries = salary[salary['playerID']=='bondsba01']['salary']

plotter(years,salaries, 'bondsba01')
```

```python
# Q2, list of players with highest salary
top_salaries = salary['salary'].sort_values(ascending = False)
indexes = [20286, 21109, 25965]

print(salary.iloc[20286])

print('\n', salary.iloc[21109])

print('\n', salary.iloc[25965])
```
